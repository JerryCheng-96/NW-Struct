```python
import pickle

id = "4NHH"
filename = "pdbs/4NHH.pdb"
struct = pp.get_structure(id, filename)

f = open("pdbs/4NHH_DSSP.dat", "rb")
dssp_4nhh = pickle.load(f)
dssp_4nhh = [dssp_4nhh[dssp_4nhh.keys()[i]] for i in range(0, len(dssp_4nhh))]

model = struct[0]
chain = model['A']
resis = [resi for resi in chain if resi.get_id()[0] == ' ']

listNVtr = [r for r in ppdb.Get_AnimoNVtr(resis)]
listNVtr_Beta = []
for i in range(0, len(listNVtr)):
    if dssp_4nhh[i][2] == 'E':
        listNVtr_Beta.append(listNVtr[i])
        
x1 = [r[0] for r in listNVtr]
y1 = [r[1] for r in listNVtr]
z1 = [r[2] for r in listNVtr]

x1_ = [r[0] for r in listNVtr_Beta]
y1_ = [r[1] for r in listNVtr_Beta]
z1_ = [r[2] for r in listNVtr_Beta]

plt.rcParams['figure.figsize'] = (10.0, 7.5)
ax = plt.subplot(111, projection='3d')
##ax.scatter(x1, y1, z1, c='g')
ax.scatter(x1_, y1_, z1_, c='g')
ax.scatter(x, y, z, c='r')
ax.set_zlim(-2.5, -1)
ax.set_zlabel("Z")
ax.set_ylabel("Y")
ax.set_xlabel("X")
```
